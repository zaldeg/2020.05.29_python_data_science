```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = 'creditcard.csv'
df = pd.read_csv(DATASET_PATH)
df
```

```python
df.info()
```

```python
class_counts = df['Class'].value_counts()
df.shape
```

```python
class_counts.plot(kind="bar")
```

```python
class_counts.plot(kind="bar", logy=True)
```

```python
V1_fraud = df.loc[(df['Class'] == 1)][['V1', 'Class']]
V1_honest = df.loc[(df['Class'] == 0)][['V1', 'Class']]
```

```python
plt.style.use('fivethirtyeight')
plt.hist(V1_fraud['V1'], bins=20, alpha=0.5, density=True, label='Class 1', color='red')
plt.hist(V1_honest['V1'], bins=20, alpha=0.5, density=True,label='Class 0', color='grey')
plt.xlabel('Class')
plt.legend(loc='upper left')
plt.show()
```
